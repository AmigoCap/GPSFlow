```python
def getFilteredDistances(df, lat, lng):
    size = df['timestampMs'].size
    distances = []
    for i in range(size - 1):
        distances.append(distance.haversineDistance(
            df[lng][i],
            df[lat][i],
            df[lng][i+1],
            df[lat][i+1]))

    distances.append(0)

    return distances
```

```python
day_df['mean_distance'] = getFilteredDistances(day_df, "lat_mean_filt", "lng_mean_filt");
```

```python
def fdistance(df, i, j) :
    return distance.haversineDistance(
        df["lng_mean_filt"][i],
        df["lat_mean_filt"][i],
        df["lng_mean_filt"][j],
        df["lat_mean_filt"][j])
```

```python
init_stay_point = []
end_stay_point = []

def findStayPoints(df, lower_limit, radius, max_outliers) :
    i = 0
    while i < df["timestampMs"].size - lower_limit :
        
        # print ("Testing " + str(i))
        # Vérifier si i est un début de stay point
        mouvement = False;
        for k in range(lower_limit) :
            if (fdistance(df, i, i + k) > radius) :
                mouvement = True
        
        if mouvement :
            # Si on est en mouvement, suivant
            i += 1
        else :
            # Si on est immobile, trouver jusqu'à quel indice
            init_stay_point.append(i)
            print ("DEBUT : " + str(i))
            outliers = max_outliers
        
            j = i + 1
            while (outliers >= 0 and j < df["timestampMs"].size) :
                if (fdistance(df, i, j) > radius) :
                    outliers -= 1  
                    #print ("Outlier detected")
                else :
                    outliers = max_outliers
                j += 1
            i = j - max_outliers
            end_stay_point.append(i)
            print ("FIN : " + str(i))
```
